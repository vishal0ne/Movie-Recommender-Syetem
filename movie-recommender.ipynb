```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_f['genres'].apply(convert)
```

```python
movies_f['genres'] = movies_f['genres'].apply(convert)
```

```python
movies_f['keywords'] = movies_f['keywords'].apply(convert)
```

```python
def convert2(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies_f['cast'].apply(convert2)
```

```python
movies_f['cast'] = movies_f['cast'].apply(convert2)
```

```python
movies_f.iloc[0].crew
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies_f['crew'] = movies_f['crew'].apply(convert3)
```

```python
movies_f['overview'] = movies_f['overview'].apply(lambda x: str(x).split())
```

```python
movies_f.head(2)
```

```python
movies_f['genres'] = movies_f['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_f['keywords'] = movies_f['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies_f['cast'] = movies_f['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies_f['crew'] = movies_f['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies_f.head()
```

```python
movies_f['tags'] = movies_f['overview'] + movies_f['genres'] + movies_f['keywords'] + movies_f['cast']+ movies_f['crew']
```

```python
movies_final = movies_f[['movie_id', 'title', 'tags']]
```

```python
movies_final.head()
```

```python
movies_final['tags'] = movies_final['tags'].apply(lambda x: " ".join(x))
```

```python
movies_final['tags'] = movies_final['tags'].apply(lambda x:x.lower())
```

Vectorization

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
movies_final['tags'] = movies_final['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(movies_final['tags']).toarray()
```

```python
vectors.shape
```

```python
vectors[0]
```

```python
cv.get_feature_names_out()
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
```

Cosine similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity[1]
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = movies_final[movies_final['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(movies_final.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
